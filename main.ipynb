```python
# Exercice 9

# 1 Créez une nouvelle colonne "title" qui extrait le titre du nom de chaque passager (par exemple, M. pour Monsieur, Mme pour Madame, etc.)

import re

pattern = r'(?:Mr|Mrs|Miss|Ms|Master|the Countess|Jonkheer|Mlle|Mme|Don|Rev|Capt|Dr|Major|Col|Sir|Lady)\.'

titles = [re.search(pattern, string).group(0) if re.search(pattern, string) else None for string in SeaCareerSet.name]
```

```python
titles_dict = {
    'Mr.': 'Mister',
    'Miss.': 'Miss',
    'Mrs.': 'Missus',
    'the Countess.': 'the Countess',
    'Jonkheer.': 'Jonkheer',
    'Master.': 'Master',
    'Ms.': 'Ms.',
    'Mlle.': 'Mademoiselle',
    'Mme.': 'Madame',
    'Don.': 'Don',
    'Rev.': 'Reverend',
    'Capt.': 'Captain',
    'Dr.': 'Doctor',
    'Major.': 'Major',
    'Col.': 'Colonel',
    'Sir.': 'Sir',
    'Lady.': 'Lady'
}
```

```python
SeaCareerSet['Title'] = pd.Series(titles).replace(titles_dict)
```

```python
# 2 Explorez la distribution des titres et analysez s'il y a des titres rares ou des erreurs

SeaCareerSet['Title'].value_counts()
```

```python
# Créez une nouvelle colonne appelée "group_age" dans le DataFrame, qui indique en fonction de l’age du passager : baby : de la naissance à 2 ans. kid : de 3 ans à 11 ans. adolescent : de 12 ans à 17 ans environ. adult : 18 ans et +

def assign_age_group(age):
    if age <= 2:
        return 'baby'
    elif age <= 11:
        return 'kid'
    elif age <= 17:
        return 'adolescent'
    else:
        return 'adult'

SeaCareerSet['group_age'] = SeaCareerSet['age'].apply(assign_age_group)
```
